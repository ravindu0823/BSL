```python
import cv2
import mediapipe as mp
import os
import numpy as np 
```

```python
def image_processed(file_path):
    hand_img = cv2.imread(file_path)
    img_rgb = cv2.cvtColor(hand_img, cv2.COLOR_BGR2RGB)

    mp_hands = mp.solutions.hands
    hands = mp_hands.Hands(static_image_mode=True,
                           max_num_hands=2,
                           min_detection_confidence=0.7)

    output = hands.process(img_rgb)
    hands.close()

    landmarks = []
    if output.multi_hand_landmarks:
        for hand_landmarks in output.multi_hand_landmarks:
            for lm in hand_landmarks.landmark:
                landmarks.extend([lm.x, lm.y, lm.z])
    else:
        return(np.zeros(126)) 

    return landmarks

def make_csv():
    mypath = '2_HAND_DATASET'
    file_name = open('../two_hand_dataset.csv', 'w')
    for each_folder in os.listdir(mypath):
        if '._' in each_folder or not os.path.isdir(os.path.join(mypath, each_folder)):
            continue

        for each_file in os.listdir(os.path.join(mypath, each_folder)):
            if '._' in each_file:
                pass
            else:
                label = each_folder
                file_loc = os.path.join(mypath, each_folder, each_file)
                data = image_processed(file_loc)
                data_string = ','.join(map(str, data)) + ',' + label + '\n'
                file_name.write(data_string)
       
    file_name.close()
    print('Data Created !!!')

if __name__ == "__main__":
    make_csv()
```
